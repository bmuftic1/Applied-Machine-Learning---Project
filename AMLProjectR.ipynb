```r
fileTrain<-'C:\\Users\\MM\\Desktop\\DIT\\AppliedMachineLearning\\Project\\trainModified.csv'
train<-read.csv(fileTrain)
```

```r
Data<-train[,c("Pclass", "Sex", "Age", "SibSp", "ParCh", "Ticket", "Fare", "Cabin", "Embarked")]
wss<-(nrow(Data)-1)*sum(apply(Data,2,var))
for (i in 2:12) {
    set.seed(1234)
    wss[i]<-sum(kmeans(Data,centers=i)$withinss)
}
plot(1:12, wss, type="b", xlab="Number of clusters", ylab="Within groups sum of squares")
```

```r
#optimum is 5
set.seed(987654321)

clusterOpt<- kmeans(train[, c("Pclass", "Sex", "Age", "SibSp", "ParCh", "Ticket", "Fare", "Cabin", "Embarked")],centers=5, nstart=25)
ord=order(clusterOpt$cluster)
data.frame(train$PassengerId[ord], clusterOpt$cluster[ord])
```
